```python
#Task - 3 - Outlier Detection:
import pandas as pd
import matplotlib.pyplot as plt

dataset_path = '/content/drive/MyDrive/Assignment1/south_asian_countries.csv'
south_asian_df = pd.read_csv(dataset_path)

# Step 1: Identify outliers for Score and GDP per Capita using the 1.5 × IQR rule
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

outliers_score = detect_outliers(south_asian_df, 'score')
outliers_gdp = detect_outliers(south_asian_df, 'Log GDP per capita') # Change to the actual column name

outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

# Step 2: Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis
plt.figure(figsize=(10, 6))

plt.scatter(south_asian_df['Log GDP per capita'], south_asian_df['score'], color='green', label='Normal')

plt.scatter(outliers['Log GDP per capita'], outliers['score'], color='orange', label='Outliers') # Change x-axis data here as well

plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('Outlier Detection: GDP per Capita vs Happiness Score')
plt.legend()
plt.grid(True)
plt.show()

# Step 3: Discuss the characteristics of the outliers
print("Outlier countries based on Score and GDP per Capita:")
print(outliers[['Country name', 'Log GDP per capita', 'score']])
```

Afghanistan stands out as an outlier with very low Log GDP per capita (0.628) and a score of 1.721, highlighting its significant economic challenges and poor social conditions caused by ongoing conflict and instability. Including Afghanistan in regional analyses can lower averages, reveal stark inequalities, and show the importance of focusing on tailored development strategies. Its outlier status suggests that using medians or removing extreme values could provide more accurate regional insights. Afghanistan’s difficulties also affect stability in the region, making development and peacebuilding efforts essential.

```python
#Task - 4 - Exploring Trends Across Metrics:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

dataset_path = '/content/drive/MyDrive/Assignment1/south_asian_countries.csv'
south_asian_df = pd.read_csv(dataset_path)

# Step 1: Calculate Pearson correlation between selected metrics and Score
freedom_corr, _ = pearsonr(south_asian_df['Freedom to make life choices'], south_asian_df['score'])
generosity_corr, _ = pearsonr(south_asian_df['Generosity'], south_asian_df['score'])

print(f"Pearson correlation between 'Freedom to Make Life Choices' and Score: {freedom_corr:.2f}")
print(f"Pearson correlation between 'Generosity' and Score: {generosity_corr:.2f}")

# Step 2: Create scatter plots with trendlines
plt.figure(figsize=(10, 6))
sns.regplot(x='Freedom to make life choices', y='score', data=south_asian_df, scatter_kws={'color': 'blue'}, line_kws={'color': 'green'})
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Happiness Score')
plt.title('Freedom to Make Life Choices vs Happiness Score')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(x='Generosity', y='score', data=south_asian_df, scatter_kws={'color': 'blue'}, line_kws={'color': 'orange'})
plt.xlabel('Generosity')
plt.ylabel('Happiness Score')
plt.title('Generosity vs Happiness Score')
plt.grid(True)
plt.show()

# Step 3: Discussion of strongest and weakest relationships
if freedom_corr > generosity_corr:
    strongest_metric = 'Freedom to Make Life Choices'
    weakest_metric = 'Generosity'
else:
    strongest_metric = 'Generosity'
    weakest_metric = 'Freedom to Make Life Choices'

print(f"\nThe strongest correlation with the Happiness Score is: {strongest_metric}")
print(f"The weakest correlation with the Happiness Score is: {weakest_metric}")
```

The data shows that Generosity has the strongest positive relationship with Happiness Score among South Asian countries, with a Pearson correlation of 0.87. The graph reveals a clear upward trend, suggesting that higher levels of generosity are strongly linked to higher happiness levels. This highlights the cultural importance of altruism and community support in the region, which likely contributes to overall well-being.

On the other hand, Freedom to Make Life Choices shows a slightly weaker correlation with Happiness Score, at 0.80. While there is still a positive relationship, the data points are more scattered, indicating that other factors, such as economic or societal constraints, might limit the impact of personal freedom on happiness in South Asia. Both factors are important, but generosity seems to play a more significant role in shaping happiness in this context.

Analysis for Step 4:

Reasons for Large Gaps:

High GDP but Low Happiness (Negative Gap):
Countries with high GDP but low happiness may have problems like wealth inequality, lack of support systems, or political problems, which reduce people's overall well-being even though the economy is growing.

Low GDP but High Happiness (Positive Gap):
Countries with low GDP but high happiness may have strong community support, good social systems, or cultural factors that help people feel happy despite having a smaller economy.

Implications:

A large positive gap means that GDP alone isn’t enough to explain happiness. These countries might have good social or health services that make people happier even without a strong economy.
A large negative gap means that even though a country has economic growth, other problems like inequality or corruption might prevent people from enjoying the benefits of that growth.
