```python
# Handling Missing values - Adding Missing Values.
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])

# Introduce missing values randomly
np.random.seed(42)  # Set random seed for reproducibility
mask = np.random.rand(*iris_df.shape) < 0.1  # 10% missing values
iris_df[mask] = np.nan

# Print the number of missing values per column
print("Missing Values in Iris Dataset:")
print(iris_df.isnull().sum())
```

```python
# Handling Missing values - Techniques for Filling Missing Values.
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])

# Introduce missing values randomly
np.random.seed(42)  # Set random seed for reproducibility
mask = np.random.rand(*iris_df.shape) < 0.1  # 10% missing values
iris_df[mask] = np.nan

# Filling missing values with different strategies
iris_df_ffill = iris_df.ffill()  # Forward fill
iris_df_mean = iris_df.fillna(iris_df.mean())  # Fill with mean
iris_df_median = iris_df.fillna(iris_df.median())  # Fill with median
iris_df_zero = iris_df.fillna(0)  # Fill with 0

# Expand iris_df with filled columns
iris_df_expanded = pd.concat([
    iris_df,
    iris_df_ffill.add_suffix('_ffill'),
    iris_df_mean.add_suffix('_mean'),
    iris_df_median.add_suffix('_median'),
    iris_df_zero.add_suffix('_zero')
], axis=1)

# Display the head of the expanded DataFrame
print("\nDataset after Filling Missing Values:")
print(iris_df_expanded.head())
```

```python
# - Some Common Operations on Data Cleaning.
import pandas as pd

# ---------------------------------------------------------
# ----------- Trimming Whitespaces: ----------------------
# ---------------------------------------------------------
df = pd.DataFrame({'Name': ['Alice ', 'Bob '], 'Age': [25, 30]})
df['Name'] = df['Name'].str.strip()
print("DataFrame after trimming whitespaces:")
print(df)

# ---------------------------------------------------------
# ------------- Changing Datatype: ------------------------
# ---------------------------------------------------------
df = pd.DataFrame({'Age': ['25', '30', '35']})
# Change 'Age' column data type to integer
df['Age'] = df['Age'].astype(int)
print("\nDataFrame after changing datatype:")
print(df)

# ---------------------------------------------------------
# ------------- Renaming Columns: -------------------------
# ---------------------------------------------------------
df = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
# Rename columns
df = df.rename(columns={'Name': 'Full Name', 'Age': 'Years'})
print("\nDataFrame after renaming columns:")
print(df)

# ---------------------------------------------------------
# -------------- Removing Duplicates: ---------------------
# ---------------------------------------------------------
df = pd.DataFrame({'Name': ['Alice', 'Bob', 'Alice'], 'Age': [25, 30, 25]})
# Remove duplicate rows
df = df.drop_duplicates()
print("\nDataFrame after removing duplicates:")
print(df)
```

```python
# DataFrame Reshaping - Pivot and Melt.
import pandas as pd

# ------------------ Pivoting ----------------------------------------
# Sample DataFrame
data = {'Date': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
        'City': ['Kathmandu', 'Pokhara', 'Kathmandu', 'Pokhara'],
        'Temperature': [15, 18, 16, 19]}

df = pd.DataFrame(data)

# Pivot: Reshape data to show cities as columns
pivoted_df = df.pivot(index='Date', columns='City', values='Temperature')
print("Pivoted DataFrame:")
print(pivoted_df)

# --------------------------- Melting -----------------------------------------
# Melt: Convert wide data back to long format
melted_df = pd.melt(pivoted_df.reset_index(), id_vars=['Date'],
                    var_name='City', value_name='Temperature')

print("\nMelted DataFrame:")
print(melted_df)
```

```python
# Data Transformation - Min-Max Scaling.
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])

# Min-Max Scaling using Pandas
iris_minmax_scaled = (iris_df - iris_df.min()) / (iris_df.max() - iris_df.min())

# Display the original and min-max scaled DataFrames
print("Original Iris DataFrame:")
print(iris_df.head())

print("\nMin-Max Scaled Iris DataFrame:")
print(iris_minmax_scaled.head())  # Display scaled data
```

```python
# Handling Categorical Variables - Ordinal or Label Encoding.
import pandas as pd

# Sample DataFrame with ordinal categories
df = pd.DataFrame({'Category': ['Low', 'Medium', 'High', 'Low', 'High']})

# Ordinal encoding using map
ordinal_mapping = {'Low': 1, 'Medium': 2, 'High': 3}
df['Category_Ordinal'] = df['Category'].map(ordinal_mapping)

# Display the DataFrame
print(df)
```

```python
# Handling Categorical Variables - One Hot Encoding.
import pandas as pd

# Sample DataFrame
df_municipalities = pd.DataFrame({'Municipality': ['Kathmandu', 'Bhaktapur', 'Lalitpur', 'Madhyapur Thimi', 'Kirtipur']})

# One-hot encoding using pd.get_dummies
one_hot_encoding = pd.get_dummies(df_municipalities['Municipality'], prefix='Municipality')

# Concatenate the one-hot encoded columns with the original DataFrame
df_encoded = pd.concat([df_municipalities, one_hot_encoding], axis=1)

# Display the result
print(df_encoded)
```

```python
# - Merging and Joining DataFrames - Concatenation.
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Row-wise concatenation
combined_rows = pd.concat([df1, df2], axis=0)
print("Row-wise concatenation:")
print(combined_rows)

# Column-wise concatenation
combined_cols = pd.concat([df1, df2], axis=1)
print("\nColumn-wise concatenation:")
print(combined_cols)
```

```python
# Merging and Joining DataFrames - Merge.
import pandas as pd

# Sample DataFrames
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Score': [85, 90, 88]})

# Inner join
inner_merged = pd.merge(df1, df2, on='ID', how='inner')
print("Inner Join:")
print(inner_merged)

# Left join
left_merged = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Join:")
print(left_merged)

# Outer join
outer_merged = pd.merge(df1, df2, on='ID', how='outer')
print("\nOuter Join:")
print(outer_merged)
```

```python
#Problem-2
import pandas as pd
# 1
dataset = pd.read_csv("/content/drive/MyDrive/Week2/Datasets-20241208T081916Z-001/Datasets/medical_students_dataset.csv")

# 2
print("\nDataFrame Info:")
print(dataset.info())

print("\nMissing values in each column:")
print(dataset.isnull().sum())

# 3
print("\nMissing values in each column:")
print(dataset.isnull().sum())

for column in dataset.columns:
    if dataset[column].dtype == 'object':  # Categorical columns
        # Filling missing values with the most frequent value (mode)
        dataset[column] = dataset[column].fillna(dataset[column].mode()[0])
        print(f"\nMissing values in {column} column filled with the mode.")
    elif dataset[column].dtype in ['int64', 'float64']:  # Numerical columns
        # Filling missing values with the mean
        dataset[column] = dataset[column].fillna(dataset[column].mean())
        print(f"\nMissing values in {column} column filled with the mean.")
    else:
        print(f"\nColumn {column} is of unsupported data type for imputation.")

print("\nMissing values after filling:")
# 4
print(dataset.isnull().sum())
```

```python
# Problem-5
import pandas as pd
import matplotlib.pyplot as plt
titanic = pd.read_csv("/content/drive/MyDrive/Week2/Datasets-20241208T081916Z-001/Datasets/Titanic-Dataset.csv")
grouped_data = titanic.groupby(['Embarked', 'Sex'])['Survived'].mean().unstack()
print("Mean Survival Rates by Gender and Port of Embarkation:")
print(grouped_data)
ports = ['C', 'Q', 'S']
male_survival = grouped_data.loc[ports, 'male']
female_survival = grouped_data.loc[ports, 'female']
x = range(len(ports))
plt.figure(figsize=(10, 6))
bar_width = 0.4
plt.bar(x, male_survival, width=bar_width, color='lightblue', label='Male', align='center')
plt.bar([p + bar_width for p in x], female_survival, width=bar_width, color='pink', label='Female', align='center')
plt.title("Mean Survival Rates by Gender and Port of Embarkation", fontsize=16)
plt.xlabel("Port of Embarkation", fontsize=12)
plt.ylabel("Mean Survival Rate", fontsize=12)
plt.xticks([p + bar_width / 2 for p in x], ports, fontsize=10)
plt.ylim(0, 1)
plt.legend()
plt.show()
```

```python
# Problem-6
import pandas as pd
import matplotlib.pyplot as plt

titanic = pd.read_csv("/content/drive/MyDrive/Week2/Datasets-20241208T081916Z-001/Datasets/Titanic-Dataset.csv")

titanic['AgeGroup'] = pd.qcut(titanic['Age'], q=5, labels=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'])

grouped_data = titanic.groupby(['Pclass', 'AgeGroup'])['Survived'].mean().unstack()
print("Mean Survival Rates by Age Group and Passenger Class:")
print(grouped_data)

plt.figure(figsize=(10, 6))
plt.imshow(grouped_data, cmap='Blues', aspect='auto')

plt.title("Survival Rates by Age Group and Passenger Class", fontsize=16)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Passenger Class (Pclass)", fontsize=12)
plt.xticks(range(len(grouped_data.columns)), grouped_data.columns, fontsize=10)
plt.yticks(range(len(grouped_data.index)), grouped_data.index, fontsize=10)
plt.colorbar(label="Survival Rate")
plt.show()
```
